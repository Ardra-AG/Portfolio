```python
def fill_outliers(df):
       for i in df.select_dtypes(include=["int64","float64"]).columns:
              q1=df[i].quantile(0.25)
              q3=df[i].quantile(0.75)
              IQR=q3-q1

              lower=q1-1.5*IQR
              upper=q3+1.5*IQR

              median=df[i].median()
              df[i]=df[i].apply(lambda x:median if x < lower or x> upper else x)
       return(df)
df=fill_outliers(df)
```

```python
def check_outliers(df):
       for i in df.select_dtypes(include=["int64","float64"]).columns:
              q1=df[i].quantile(0.25)
              q3=df[i].quantile(0.75)
              IQR=q3-q1

              lower=q1-1.5*IQR
              upper=q3+1.5*IQR

              out_left=df[(df[i] < lower )| (df[i] > upper)]
              print(f'column: {i}')
              print(f'lower_bound: {lower}, upper_bound: {upper}')
              print(f' number of remaining outliers : {len(out_left)}')
check_outliers(df)
```

```python
import pandas as pd
from imblearn.over_sampling import SMOTE
from collections import Counter

X = df.drop(['Churn'], axis=1)
y = df['Churn']


print("Class distribution before balancing:", Counter(y))


smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)


print("Class distribution after balancing:", Counter(y_resampled))

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X_resampled,y_resampled,test_size=0.30,random_state=42)
x_train
```

```python
from sklearn.preprocessing import StandardScaler
scal=StandardScaler()
scal.fit(x_train)
x_train=scal.transform(x_train)
x_test=scal.transform(x_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report,ConfusionMatrixDisplay

knn=KNeighborsClassifier(n_neighbors=7)
nab=BernoulliNB()
svm=SVC()
lst=[knn,nab,svm]
```

```python
for i in list:
  print(i,"Model")
  i.fit(x_train,y_train)
  y_pred=i.predict(x_test)
  print("Confusion Matrix :",confusion_matrix(y_test,y_pred))
  print("Accuracy Score :",accuracy_score(y_test,y_pred))
  print("Classification Report:",classification_report(y_test,y_pred))
  ConfusionMatrixDisplay.from_predictions(y_test, y_pred)
  plt.title("Confusion Matrix ")
  plt.show()
```
